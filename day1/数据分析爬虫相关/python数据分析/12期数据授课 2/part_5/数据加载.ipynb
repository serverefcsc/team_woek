- 读取sqlite3文件


导包

```python
import sqlite3
```

- 连接数据库  
    sqlite3.connect('dbpath')
    
- 读取table内容  
    pd.read_sql("SQL语句", con)
    
- 写入数据库文件
    df对象.to_sql(name,connection):将df中的数据值写入数据库中，表名为name
    
- 操作数据库
    connection.execute(SQL语句)

```python
#连接数据库，获取连接对象
conn = sqlite3.connect('./data-07/weather_2012.sqlite')
```

```python
#读取库表中的数据值
pd.read_sql('select * from weather_2012',conn)
```

```python
#将一个df中的数据值写入存储到db
df123.to_sql('df123',conn)
```

```python
#验证
pd.read_sql('select * from df123',conn)
```

```python
#执行sql语句
conn.execute('insert into df123 values(4,"aaa","bbb","ccc")')
```
